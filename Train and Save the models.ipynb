```python
import pandas as pd
```

```python
data = pd.read_parquet("wikipedia_simple_dump_200k.parquet")
```

```python
raw_documents = data["text"].to_list()
```

```python
len(raw_documents)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
stop_words = set(stopwords.words("english"))
def tokenize_clean(document):
    return [w.lower() for w in word_tokenize(document) if w.lower() not in stop_words and w.isalpha()]
```

```python
cleaned_corpus = []
counter = 0
for doc in raw_documents:
    cleaned_corpus.append(tokenize_clean(doc))

    if counter % 1000 == 0:
        print(counter)
    counter += 1    
```

```python
corpus = [["Rober clive is a soldier"], ["british ruled india for 200 years"]]
```
